```python
def deg2Rad(deg):
    rad = (deg/180.0)*m.pi
    return rad

def getDist(x0, y0, x1, y1):
    dist = m.sqrt((x0 - x1)**2 + (y0 - y1)**2)
    return dist

def collideTarget(bx, by, bRad, tx, ty, tw, th):
    collision = False
    if ((bx >= tx-tw/2-bRad) and (bx <= (tx+tw/2+bRad))) and ((by >= ty-th/2-bRad) and (by <= (ty+th/2+bRad))):
        collision = True
    return collision

def collideBomb(bx, by, bRad, Bx, By, Brad):
    collision = False
    myDist = getDist(bx, by, Bx, By)
    if (myDist < (bRad + Brad)):
        collision = True
    return collision
```

```python
# --- Check For close Stuff
def distance(items, turret):
    closest_item = None
    closest_index = 0
    for i in range(len(items)):
        item = items[i]
        iXY = item.x, item.y
        tXY = turret.x, turret.y
        if(closest_item == None):
            closest_item = m.dist(iXY, tXY)
            closest_index = i
        elif(closest_item > m.dist(iXY, tXY)):
            closest_item = m.dist(iXY, tXY)
            closest_index = i

    return items[closest_index]
```

My neural network class Needs to work as so: It needs to be called in the main program and have the number of neurons, layers, and learning_rate set.
